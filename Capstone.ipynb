```python
import pandas as pd 

df = pd.read_csv("./RecruitingClasses2019_2022.csv")
df2 = pd.read_csv("./cfb22Cleaned.csv")
# cleaning this csv so that it will show Offensive, Defensive, and Special players instead of unique positions 
df['POS'] = df['POS'].replace(['WDE', 'SDE'], 'DE')
df['POS'] = df['POS'].replace(['PRO', 'DUAL'], 'QB')
df['POS'] = df['POS'].replace(['OT', 'IOL', 'OG', 'C', 'OC'], 'OL')
df['POS'] = df['POS'].replace(['APB'], 'RB')
df['POS'] = df['POS'].replace(['DE', 'DT'], 'DL')
df['POS'] = df['POS'].replace(['CB', 'S'], 'DB')
df['POS'] = df['POS'].replace(['QB', 'WR', 'FB', 'RB', 'OL', 'TE'], 'OFFENSE')
df['POS'] = df['POS'].replace(['DL', 'Edge', 'OLB', 'ILB', 'LB', 'DB'], 'DEFENSE')
df['POS'] = df['POS'].replace(['P', 'K', 'LS', 'ATH'], 'SPECIAL')


df.to_csv('./CleanedRecruitingClasses19-22.csv')
df
```
